#seaborn is built on top of matplotlib it is also used in data analysis and data vissualization

seaborn provides you to create different kinds plots

seaborn now widely used tools compare with matplotlib

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset('tips')
```

```python
pf=pd.DataFrame(data)
pf.head()
```

```python
pf['sex']=pf['sex'].apply(lambda x:0 if x=='Female' else 1)
```

```python
pf.head()
```

```python
men_women=pf.groupby('sex')['total_bill'].sum()
print(men_women)
```

```python
men_women.plot(kind='bar')
```

```python
#correlation
cor=pf[['total_bill','tip','size']].corr()
```

```python
sns.heatmap(cor,cmap="Blues",cbar=True,annot=True)
plt.show()
```

```python
#it is used for identify the outliers in our dataset
sns.boxplot(x='total_bill',data=pf)
plt.show()
```

```python
#pairplot is used to check how one feature is related to  others
sns.pairplot(pf)
plt.show()
```

```python
#voilin plot- it is ued to check outliers
sns.violinplot(x='total_bill',data=pf)
plt.show()
```

```python
#scatterplot is used to check the distribution of data
sns.scatterplot(x="total_bill",y='tip',data=pf)
plt.show()
```

```python
#kernel density estimatimation
sns.kdeplot(x='total_bill',data=pf,fill=True)
plt.show()
```

```python
sns.histplot(pf['total_bill'],bins=10,kde=True)
plt.show()
```

```python
sns.countplot(x='size',data=pf)
plt.show()
```

```python
sns.lineplot(x='size',y='total_bill',data=pf)
plt.show()
```
